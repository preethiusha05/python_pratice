```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
df[df['SalePrice'] > 200000]
```

```python
top_prices = df.sort_values(by='SalePrice', ascending=False)
top_prices[['Id', 'Neighborhood', 'GrLivArea', 'SalePrice']].head()
```

```python
price_array = np.array(df['SalePrice'])
area_array = np.array(df['GrLivArea'])
```

```python
np.corrcoef(price_array, area_array)
```
