```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("food_delivery_orders.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df.head()
```

```python
df["Total_Price"] = df["Quantity"] * df["Price"]

df.head()
```

```python
df.sort_values(by="Total_Price", ascending=False).head()
```

```python
df[df["City"] == "Chennai"]
```

```python
df["Payment_Method"].value_counts()
```

```python
df.groupby("Restaurant")["Total_Price"].sum()
```

```python
df["Order_Category"] = df["Total_Price"].apply(
    lambda x: "High" if x > 500 else "Medium" if x > 200 else "Low")
df.head()
```

```python
df.rename(columns={"Food_Item":"Item"}, inplace=True)
df.head()
```

```python
df["City"] = df["City"].str.upper()
df.head()
```

```python
df[["Quantity","Price","Delivery_Time"]].corr()
```

```python
df["Restaurant"]=df["Restaurant"].str.lower()
df.head()
```

```python
quantity_array = np.array(df["Quantity"])

print(quantity_array)
```

```python
print("Average Quantity:", np.mean(quantity_array))
print("Max Quantity:", np.max(quantity_array))
print("Min Quantity:", np.min(quantity_array))
```

```python
new_quantity = quantity_array + 2

print(new_quantity[:10])
```

```python
matrix = np.array([[1,2],[3,4]])
print("Determinant:", np.linalg.det(matrix))
print("Inverse:", np.linalg.inv(matrix))
print("Eigenvalues:", np.linalg.eig(matrix)[0])
```
