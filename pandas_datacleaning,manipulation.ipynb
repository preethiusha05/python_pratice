```python
import pandas as pd
import numpy as np
df = pd.read_csv("Mall_Customers.csv")
print("Dataset Loaded")
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
gender_count = df["Gender"].value_counts()
print(gender_count)
```

```python
print(df.isnull().sum())
print("\nMissing Values:")
```

```python
df_drop = df.dropna()
print("\nAfter dropna():")
print(df_drop.head())
```

```python
df.rename(columns={
    "Annual Income (k$)": "Annual_Income",
    "Spending Score (1-100)": "Spending_Score"
}, inplace=True)

print(df.head())
```

```python
age_filter = df[df["Age"] > 30]
print(age_filter.head())
```

```python
df_fill = df.fillna(0)
print("\nAfter fillna():")
print(df_fill.head())
```

```python
sorted_df = df.sort_values(by="Annual Income (k$)", ascending=False)
print("\nSorted Data:")
print(sorted_df.head())
```

```python
sorted_df = df.sort_values(by="Spending_Score", ascending=False)
print(sorted_df.head())
```

```python
group = df.groupby("Gender")["Spending_Score"].mean()
print(group)
```

```python
filtered_df = df[df["Age"] > 30]
print("\nFiltered Data (Age > 30):")
print(filtered_df.head())
```

```python
df["Income_after_tax"] = df["Annual Income (k$)"] * 0.9
print("\nNew Column Added:")
print(df.head())
```

```python
df["Income_After_Tax"] = df["Annual_Income"] * 0.9
print(df.head())
```

```python
group_data = df.groupby("Gender")["Spending Score (1-100)"].mean()
print("\nAverage Spending Score by Gender:")
print(group_data)
```
