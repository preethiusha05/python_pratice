## Scatter Plot Examples

```python
#Example 1 – Study vs Marks
import matplotlib.pyplot as plt

study = [1,2,3,4,5]
marks = [50,55,65,70,90]

plt.scatter(study,marks,color="red")

plt.title("Study Hours vs Marks")

plt.show()
```

```python
#Example - 2 Height vs Weight
height = [150,160,165,170,175]
weight = [50,60,65,70,75]

plt.scatter(height,weight,color="green")

plt.title("Height vs Weight")

plt.show()
```

```python
#Example - 3 Advertising vs Sales
import matplotlib.pyplot as plt

advertising = [100,200,300,400,500]
sales = [10,20,35,45,60]

plt.scatter(advertising, sales, color="blue", marker="o")

plt.title("Advertising vs Sales")
plt.xlabel("Advertising Budget")
plt.ylabel("Sales")


plt.show()
```

```python
#Example - 4 Experience vs Salary
import matplotlib.pyplot as plt

experience = [1,2,3,5,7,10]
salary = [20000,25000,30000,40000,50000,65000]

plt.scatter(experience, salary, color="orange", marker="^", s=120)

plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.grid(True)

plt.show()
```

```python
#Example - 5 Discount Percentage vs Sales
import matplotlib.pyplot as plt

discount = [5,10,15,20,25,30]
sales = [200,240,300,350,420,480]

plt.scatter(discount, sales, color="purple", marker="*", s=120)

plt.title("Discount Percentage vs Sales")
plt.xlabel("Discount (%)")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

```python
#Example 6 Advertising Budget vs Sales
import matplotlib.pyplot as plt

advertising_budget = [100,200,300,400,500,600]
sales = [150,220,300,380,450,520]

plt.scatter(advertising_budget, sales, color="blue", marker="o", s=120)

plt.title("Advertising Budget vs Sales")
plt.xlabel("Advertising Budget")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```
