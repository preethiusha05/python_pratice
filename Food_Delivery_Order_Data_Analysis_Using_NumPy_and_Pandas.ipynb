```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("order_history_kaggle_data.csv")
print(df.head())
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Order Placed At"] = pd.to_datetime(df["Order Placed At"])
```

```python
df["Order Placed At"].head()
```

```python
print(df["Order Placed At"].dtype)
```

```python
df["Items in order"] = pd.to_numeric(df["Items in order"], errors="coerce")
df["Revenue_per_item"] = df["Total"] / df["Items in order"]
print(df[["Total","Items in order","Revenue_per_item"]].head())
```

```python
df.sort_values(by="Total", ascending=False).head()
```

```python
df["City"].value_counts()
```

```python
df[df["City"].str.strip() == "Delhi NCR"]
```

```python
df.groupby("City")["Total"].sum()
```

```python
df["Order Status"].value_counts()
```

```python
df.groupby("Restaurant name")["Total"].sum().sort_values(ascending=False).head(10)
```

```python
df["Order_Category"] = df["Total"].apply(lambda x: "High" if x>500 else "Medium" if x>200 else "Low")
df[["Total","Order_Category"]].head()
```

```python
df.rename(columns={"Restaurant name":"Restaurant"}, inplace=True)
```

```python
df["City"] = df["City"].str.upper()
print(df["City"].str.upper())
```

```python
df["City"] = df["City"].str.lower()
print(df["City"].str.lower())
```

```python
top_restaurants = df.groupby("Restaurant")["Total"].sum().sort_values(ascending=False).head(10)
print(top_restaurants)
```

```python
df["Distance"].mean()
df.groupby("City")["Distance"].mean()
df[["Distance","Total"]].corr()
```

```python
df["Order Placed At"] = pd.to_datetime(df["Order Placed At"], errors="coerce")
df["Month"] = df["Order Placed At"].dt.month
df["Month_Name"] = df["Order Placed At"].dt.month_name()
```

```python
monthly_orders = df.groupby("Month_Name")["Order ID"].count().sort_values(ascending=False)

print(monthly_orders)
```

```python
df["Day"] = df["Order Placed At"].dt.day
df["Weekday"] = df["Order Placed At"].dt.day_name()
df["Hour"] = df["Order Placed At"].dt.hour
```

```python
df["Weekday"].value_counts()
```

```python
df["Hour"].value_counts().sort_index()
```

```python
distance_array = np.array(df["Distance"])
print("Average Distance:", np.mean(distance_array))
print("Max Distance:", np.max(distance_array))
print("Min Distance:", np.min(distance_array))
```

```python
new_distance = distance_array + 2
```

```python
matrix = np.array([[2,3],[4,5]])
print("Determinant:",np.linalg.det(matrix))
print("Inverse:",np.linalg.inv(matrix))
print("Eigenvalues:",np.linalg.eig(matrix)[0])
```
