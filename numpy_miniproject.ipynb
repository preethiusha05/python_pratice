```python
import numpy as np
np.random.seed(10)
sales = np.random.randint(1000, 10000, size=(50, 4))
print("Sales Data:\n", sales)
```

```python
store_ids = np.arange(1, 51)
data = np.column_stack((store_ids, sales))
print("Combined Data:\n", data)
```

```python
prodA, prodB, prodC, prodD = np.split(sales, 4, axis=1)
prodA = prodA.flatten()
prodB = prodB.flatten()
prodC = prodC.flatten()
prodD = prodD.flatten()
```

```python
def mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    return values[np.argmax(counts)]

print("Mode Product A:", mode(prodA))
```

```python
weights = np.array([0.3, 0.25, 0.25, 0.2])
weighted_sales = np.dot(sales, weights)
print("Weighted Sales:\n", weighted_sales)
```

```python
normalized_sales = (weighted_sales - np.min(weighted_sales)) / (np.max(weighted_sales) - np.min(weighted_sales))

print("Normalized Sales:\n", normalized_sales)
```

```python
rank_indices = np.argsort(-weighted_sales)

print("Top 5 Stores:")
print(store_ids[rank_indices[:5]])
```

```python
X = np.random.randint(100, 1000, size=(100, 3))
true_weights = np.array([50, 20, 5])  
noise = np.random.randn(100) * 1000
y = X @ true_weights + noise 
X_T = X.T
product = X_T @ X  
```

```python
det_value = np.linalg.det(product)
print("Determinant:", det_value)
```

```python
inverse_matrix = np.linalg.inv(product)
```

```python
beta = np.linalg.inv(X_T @ X) @ X_T @ y
print("Predicted Weights:", beta)
```

```python
y_pred = X @ beta
mse = np.mean((y - y_pred) ** 2)
print("MSE:", mse)
```
