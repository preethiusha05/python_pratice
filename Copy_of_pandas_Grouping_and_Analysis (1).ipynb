```python
import pandas as pd

df = pd.read_csv("Mall_Customers.csv")
group = df.groupby("Gender")["Spending Score (1-100)"].mean()
print(group)
```

```python
income = df.groupby("Gender")["Annual Income (k$)"].mean()
print(income)
```

```python
analysis = df.groupby("Gender").agg({
    "Age": "mean",
    "Annual Income (k$)": "mean",
    "Spending Score (1-100)": "mean"
})

print(analysis)
```

```python
count = df.groupby("Gender").size()
print(count)
```

```python
max_score = df.groupby("Gender")["Spending Score (1-100)"].max()
print(max_score)
```

```python
data1 = {
    "CustomerID": [1,2],
    "Age": [25,30]
}

data2 = {
    "CustomerID": [3,4],
    "Age": [35,40]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

result = pd.concat([df1, df2])

print(result)
```

```python
data = {
    "Customer": ["A", "B", "C"],
    "Join_Date": ["2023-01-10", "2022-06-15", "2024-02-20"]
}

df = pd.DataFrame(data)
df["Join_Date"] = pd.to_datetime(df["Join_Date"])

print(df)
print(df.dtypes)
```

```python
df = pd.read_csv("Mall_Customers.csv")
df["Income_Increase"] = df["Annual Income (k$)"].apply(lambda x: x * 1.1)

print(df.head())
```

```python
df["Gender"] = df["Gender"].str.upper()

print(df.head())
```

```python
df["Gender"] = df["Gender"].str.lower()

print(df.head())
```

```python
df.rename(columns={
    "Annual Income (k$)": "Annual_Income",
    "Spending Score (1-100)": "Spending_Score"
}, inplace=True)

print(df.head())
```
