# **Bar Chart Examples**

```python
#Example 1 – Product Sales
import matplotlib.pyplot as plt
products = ["Laptop","Phone","Tablet","Watch","TV"]
sales = [120,200,90,150,95]

plt.bar(products,sales,color="skyblue")
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
#Example - 2 Website Visitors
import matplotlib.pyplot as plt
websites = ["Google","YouTube","Facebook","Instagram","Whatsapp"]
visitors = [90,80,60,95,100]

plt.bar(websites, visitors, color="purple")

plt.title("Website Visitors")
plt.xlabel("Websites")
plt.ylabel("Number of Visitors")

plt.show()
```

```python
#Example - 3 Students Marks
students = ["A","B","C","D","E"]
marks = [80,90,60,99,70]

plt.bar(students,marks,color="green")

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
#Example 4 – Monthly Revenue
months = ["Jan","Feb","Mar","Apr"]
revenue = [5000,7000,6500,8000]

plt.bar(months,revenue,color="orange")

plt.title("Monthly Revenue")
plt.ylabel("Revenue")

plt.show()
```

```python
#Example - 5 Stationary Items Sales Analysis
items = ["Pen","Pencil","Eraser","Scale"]
quantity = [50,30,20,25]

colors = ["red","blue","green","purple"]

plt.bar(items,quantity,color=colors)

plt.title("Stationary Items")

plt.show()
```

```python
# Example - 6 Population comparison
countries = ["India","USA","China","Japan"]
population = [1400,330,67,125]

plt.bar(countries,population,color="teal")
plt.xlabel("countries")
plt.ylabel("population")

plt.title("Population Comparison")
plt.show()
```
