```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,20,25,30]

plt.plot(x, y, color="blue")
plt.title("Line with Different Color")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y, linestyle=":")
plt.title("Different Line Style")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [5,10,15,20,25]
plt.plot(x, y)
plt.grid(True)
plt.title("Plot with Grid")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y, marker="o", markersize=10)
plt.title("Markers with Size")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y, marker="o")
plt.annotate("Highest Value",xy=(5,30),xytext=(2,22),arrowprops=dict(facecolor='black'))
plt.title("Annotation Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [5,10,15,20,25]
plt.plot(x, y)
plt.gca().set_facecolor("lightgray")
plt.title("Background Color Change")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y, label="Monthly Revenue", color="green")
plt.legend()
plt.title("Company Performance")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y,
         color="blue",
         linestyle=":",
         marker="o",
         markersize=8,
         label="Monthly Sales")
plt.grid(True)
plt.annotate("Peak",
             xy=(5,30),
             xytext=(2,22),
             arrowprops=dict(facecolor='black'))
plt.gca().set_facecolor("lightyellow")
plt.title("Matplotlib Features Example")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
