# Histogram Examples

```python
#Example 1 – Age Distribution
import matplotlib.pyplot as plt
ages = [21,44,30,18,26,42,22,37,25,50]

plt.hist(ages,bins=5)

plt.title("Age Distribution")

plt.show()
```

```python
#Example 2 – Marks Distribution
import matplotlib.pyplot as plt
marks = [60,70,75,22,55,90,95,70,75,80]

plt.hist(marks,bins=6,color="green")

plt.title("Marks Distribution")

plt.show()
```

```python
#Example 3 – Salary Distribution
salary = [20000,35000,45000,10000,55000,70000]

plt.hist(salary,bins=5,color="orange")

plt.title("Salary Distribution")

plt.show()
```

```python
#Example - 4 Temperature Distribution
import matplotlib.pyplot as plt
temperature = [28,30,29,31,32,30,29,28,33,34]

plt.hist(temperature, bins=5, color="skyblue")

plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")

plt.show()
```

```python
#example - 5 Daily Sales Distribution
import matplotlib.pyplot as plt

daily_sales = [200,220,250,300,280,260,240,230,270,290,310,320]

plt.hist(daily_sales, bins=5, color="lightgreen", edgecolor="black")

plt.title("Daily Sales Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")


plt.show()
```

```python
#Example - 6 Monthly Sales Performance
import matplotlib.pyplot as plt

monthly_sales = [1500,1800,2000,2200,2100,2300,2500,2400,2600,2700]

plt.hist(monthly_sales, bins=6, color="pink", edgecolor="black")

plt.title("Monthly Sales Performance")
plt.xlabel("Sales Value")
plt.ylabel("Frequency")
plt.show()
```
