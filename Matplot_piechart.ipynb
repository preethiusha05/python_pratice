# Pie Chart Examples

```python
#Example 1 – Programming Language Usage
import matplotlib.pyplot as plt
labels = ["Python","Java","C++","JavaScript"]
sizes = [90,80,70,85]

plt.pie(sizes,labels=labels,autopct="%1.1f%%")

plt.title("Programming Language Usage")

plt.show()
```

```python
#Example 2 – Market Share
companies = ["A","B","C","D"]
share = [35,25,20,20]

plt.pie(share,labels=companies,autopct="%1.0f%%")

plt.title("Company Market Share")

plt.show()
```

```python
#Example - 3 Explode Pie Chart

labels = ["Food","Transport","Shopping","Bills"]
expense = [40,20,25,15]

explode = [0.1,0,0,0]

plt.pie(expense,labels=labels,explode=explode,autopct="%1.1f%%")

plt.title("Monthly Expenses")

plt.show()
```

```python
#Example - 4 Time Spent on Activities

import matplotlib.pyplot as plt
activities = ["Study","Sleep","Exercise","Entertainment"]
hours = [6,5,2,4]

plt.pie(hours, labels=activities, autopct="%1.1f%%", startangle=90)

plt.title("Daily Activities Distribution")

plt.show()
```

```python
#Example - 5 Product Sales Distribution
import matplotlib.pyplot as plt
products = ["washing machine","Mobile","Tablet","Smart Watch"]
sales = [300,450,150,200]

plt.pie(sales,labels=products,autopct="%1.1f%%")

plt.title("Product Sales Distribution")

plt.show()
```

```python
# Example - 6 Customer Preference in Jewellery
import matplotlib.pyplot as plt

types = ["Necklace", "Rings", "Bracelets", "Earrings"]
customers = [30, 35, 15, 20]

explode = [0, 0.1, 0, 0]

plt.pie(customers,labels=types,autopct="%1.1f%%")

plt.title("Customer Preference in Jewellery")

plt.show()
```
